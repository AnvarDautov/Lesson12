```python
import os
import pandas as pd
import matplotlib.pyplot as plt
root = os.getcwd()
filename = os.path.join(root, 'company_sales_data.csv')
```

```python
data = pd.read_csv(filename)
data.head()
```

```python
months = data['month_number'].tolist()
facecream = data['facecream'].tolist()
facewash = data['facewash'].tolist()
toothpaste = data['toothpaste'].tolist()
bathingsoap = data['bathingsoap'].tolist()
shampoo = data['shampoo'].tolist()
moisturizer = data['moisturizer'].tolist()
```

## 1. Данные о продажах:

```python
fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(1,1,1)
ax.plot(months,facecream, label = 'Крем для лица')
ax.plot(months,facewash, label = 'Средство для умывания')
ax.plot(months,toothpaste, label = 'Зубная паста')
ax.plot(months,bathingsoap, label = 'Мыло для купания')
ax.plot(months,shampoo, label = 'Шампунь')
ax.plot(months,moisturizer, label = 'Увлажняющий крем')
plt.title('Объемы продаж по месяцам')
plt.xlabel('Месяц')
plt.ylabel('Объем продаж', labelpad=15);
plt.xticks(months)
plt.yticks([i *1000 for i in range(16)])
plt.legend(loc = 'upper left')
plt.grid(linewidth = 1, linestyle = '--')
plt.show()
```
